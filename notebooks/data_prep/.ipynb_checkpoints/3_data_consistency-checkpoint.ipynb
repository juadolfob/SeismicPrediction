```python
#%matplotlib widget
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = pd.read_csv('../../data/DATA_1.csv', delimiter=',')
```

```python
df_2=df.loc[df.Magnitud > 3,:]

columns = df_2.columns.values
columns[4] = 'Fecha'
columns[5] = 'Hora'
df_2.columns  = columns
```

```python
df_magnitud_count = df_2[["Magnitud"]].groupby(["Magnitud"]).size().reset_index(name='counts')
df_magnitud_count.plot(x="Magnitud",y="counts")
```

<br>
<h2> b - value analysis </h2>

<p>
The magnitude count is at its peak at 3.7, thus we will choose 3.7 as the cut-off magnitude.
</p><br>

```python
df_2=df_2.loc[df_2.Magnitud >= 3.8,:]
df_2=df_2.loc[df_2.Fecha > "1974-00-00",:]
df_2.reset_index(drop=True,inplace=True)

df_magnitud_count = df_2[["Magnitud"]].groupby(["Magnitud"]).size().reset_index(name='counts')
df_magnitud_count.plot(x="Magnitud",y="counts")
```

```python
df_2.to_csv("../../data/DATA_2.csv", index=False)
```
