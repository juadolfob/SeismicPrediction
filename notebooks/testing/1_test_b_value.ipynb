### 1.1 Data Preparation

Changes made to the data for better fit our intended use.

| Modification | Description | Justification |
| ----------- | ----------- | ----------- |
| CUT-OFF MAGNITUDE | Seismic events below cut-off magnitude are removed | This ensures that misleading and incomplete information is not considered to determine the trends in seismic indicators. |
| MINIMUM DATE FOR EQUAL VARIANCE | Rows containing a value of "Date" lesser than 1974 are dropped. | There is a considerable difference in variance in data before and after 1974. |
| ONLY FLOAT MAGNITUDES | Rows containing Non-float or null values are turned into float or dropped respectively. | Non-float values of "Magnitude" values cannot be processed. |
| DROP REDUNDANT FEATURES | Columns containing redundant data are dropped. | Redundancy is unwanted. |



```python
old_df_seismic = pd.read_csv('../../data/DATA_0.csv', delimiter=',')
df_seismic = pd.read_csv('../../data/DATA_2.csv', delimiter=',')
```

```python
df_map_world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
df_map_mexico = df_map_world.loc[df_map_world.name == "Mexico",:]
```

```python
seismic_geometry = [Point(lat,lng) for lat,lng in zip(df_seismic["Longitud"],df_seismic["Latitud"])]
gdf_seismic_geometry = GeoDataFrame(df_seismic, geometry=seismic_geometry)
gdf_seismic_geometry.plot(ax=df_map_mexico.plot(figsize=(20, 12)), marker='o', color='red', markersize=((df_seismic.Magnitud-2)/3.5).pow(10))
```
